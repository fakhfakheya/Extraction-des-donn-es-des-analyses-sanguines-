```python
import pandas as pd

colonnes_sans_point_zero = [
    'neutrophiles%',
    'eosinophiles%',
    'basophiles%',
    'lymphocytes%',
    'monocytes%',
]

# Fonction pour vérifier si un nombre a plus de 2 chiffres après la virgule
def has_more_than_two_decimals(x):
    try:
        decimal_part = str(x).split(".")[1]
        return len(decimal_part) > 2
    except IndexError:
        return False  # Pas de partie décimale
    except:
        return False  # En cas d'erreur (valeurs non numériques etc.)

# On filtre les lignes qui n'ont PAS de colonnes avec + de 2 chiffres après la virgule
mask = df[colonnes_sans_point_zero].applymap(has_more_than_two_decimals).any(axis=1)

# Supprimer ces lignes et remplacer df
df = df[~mask].reset_index(drop=True)

print(df)
```

```python
import re
import unicodedata
import cv2
import pytesseract
from difflib import get_close_matches

# Configuration Tesseract
pytesseract.pytesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
image_path = r"C:\Users\asus\Desktop\generation2\generated_reports\rapport_1.png"

# Chargement et prétraitement image
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# OCR avec Tesseract
data = pytesseract.image_to_data(gray, output_type=pytesseract.Output.DICT, lang='fra')
n_boxes = len(data['text'])

# Unités valides
unites_valides = {
    "g/dL", "pg", "µm³", "um³", "fL", "%", "10^3/uL", "10^6/uL",
    "mm³", "10*3/uL", "10*6/uL", "/mm³", "10*3/mm³", "10*6/mm³", "/ml", "g/dl", "Pg",
    "10*6/µl", "/µl", "/ul", "millions/mm", "M/mm³", "femtolites","10⁶/mm³", "10p6/mm³", "10p3/mm³", "10 p 6/mm³", "10 p 3/mm³"
}

# Champs médicaux cibles
champs_cibles = [
    "leucocytes", "hématies", "plaquettes", "hémoglobine", "hématocrite",
    "neutrophiles", "lymphocytes", "eosinophiles", "basophiles", "monocytes",
    "globules rouges", "globules blancs", "hematologie", "vgm", "ccmh", "tcmh", "tgmh", "rdw", "vpm"
]
champs_sans_accents = [re.sub(r'[^a-z0-9\s]', '', unicodedata.normalize('NFD', c).lower()) for c in champs_cibles]

# Champs à extraire uniquement 2ème valeur numérique sans unité
champs_double_valeur = {"neutrophiles", "eosinophiles", "basophiles", "lymphocytes", "monocytes"}

def est_unite_complexe(mot):
    pattern = r'^\d+p\d+\/mm[3³]$'
    pattern_general = r'^\d+[a-z]+\d+\/[a-z0-9/³*^]+$'
    return bool(re.match(pattern, mot.lower()) or re.match(pattern_general, mot.lower()))

def separer_chiffres_unites(texte):
    mots = texte.split()
    mots_corriges = []
    for mot in mots:
        if est_unite_complexe(mot):
            mots_corriges.append(mot)
        else:
            mot_sep = re.sub(r'(\d)([a-zA-Zµ/%])', r'\1 \2', mot)
            mot_sep = re.sub(r'([a-zA-Zµ/%])(\d)', r'\1 \2', mot_sep)
            mots_corriges.append(mot_sep)
    return ' '.join(mots_corriges)

def normaliser(mot):
    mot = unicodedata.normalize('NFD', mot)
    return ''.join(c for c in mot if unicodedata.category(c) != 'Mn').lower()

def nettoyer_et_enlever_accents(texte):
    texte = unicodedata.normalize('NFD', texte.lower())
    return re.sub(r'[^a-z0-9\s]', '', ''.join(c for c in texte if unicodedata.category(c) != 'Mn'))

def corriger_mot_ocr(mot):
    mot = mot.replace("'", "").replace("`", "")
    mot_clean = nettoyer_et_enlever_accents(mot)
    match = get_close_matches(mot_clean, champs_sans_accents, n=1, cutoff=0.7)
    return champs_cibles[champs_sans_accents.index(match[0])] if match else mot

def corriger_unite_intelligente(mot):
    mot_normalise = normaliser(mot)
    correspondances = get_close_matches(mot_normalise, [normaliser(u) for u in unites_valides], n=1, cutoff=0.6)
    if correspondances:
        index = [normaliser(u) for u in unites_valides].index(correspondances[0])
        return list(unites_valides)[index]
    return mot

def corriger_ligne_intelligente(ligne):
    ligne = separer_chiffres_unites(ligne)
    mots = ligne.split()
    mots_corriges = []
    for mot in mots:
        mot_corrige = corriger_mot_ocr(mot)
        if mot_corrige == mot:
            mot_corrige = corriger_unite_intelligente(mot)
        mots_corriges.append(mot_corrige)
    return ' '.join(mots_corriges)

# Regroupement par ligne
lines = {}
tolerance_y = 15
for i in range(n_boxes):
    mot = data['text'][i].strip()
    if mot == '' or int(data['conf'][i]) < 0:
        continue
    y, x = data['top'][i], data['left'][i]
    found = False
    for key in lines:
        if abs(key - y) <= tolerance_y:
            lines[key].append({'word': mot, 'x': x})
            found = True
            break
    if not found:
        lines[y] = [{'word': mot, 'x': x}]

lignes_corrigees = []
for y in sorted(lines):
    mots_tries = sorted(lines[y], key=lambda m: m['x'])
    ligne = " ".join(m['word'] for m in mots_tries)
    ligne_corrigee = corriger_ligne_intelligente(ligne)
    lignes_corrigees.append(ligne_corrigee)

# Extraction des valeurs demandées, uniquement 2ème valeur numérique sans unité pour champs doubles,
# sinon 1ère valeur numérique sans unité
def extraire_valeurs_sans_unite(champ, lignes):
    champ_norm = nettoyer_et_enlever_accents(champ.lower())
    for ligne in lignes:
        ligne_norm = nettoyer_et_enlever_accents(ligne.lower())
        if champ_norm in ligne_norm:
            if champ in champs_double_valeur:
                valeurs = re.findall(r'[\d.,]+', ligne)
                if len(valeurs) >= 2:
                    val = valeurs[1].replace(',', '.')
                    return [val]
            else:
                match_val = re.search(r'[\d.,]+', ligne)
                if match_val:
                    val = match_val.group(0).replace(',', '.')
                    return [val]
    return None


# Questions à poser
questions = {
    "globules rouges": "Quelle est la valeur des globules rouges ?",
    "hématocrite": "Quelle est la valeur de l'hématocrite ?",
    "hémoglobine": "Quelle est la valeur de l'hémoglobine ?",
    "leucocytes": "Quelle est la valeur des leucocytes ?",
    "neutrophiles": "Quelle est la valeur des neutrophiles ?",
    "eosinophiles": "Quelle est la valeur des eosinophiles ?",
    "basophiles": "Quelle est la valeur des basophiles ?",
    "lymphocytes": "Quelle est la valeur des lymphocytes ?",
    "monocytes": "Quelle est la valeur des monocytes ?",
    "plaquettes": "Quelle est la valeur des plaquettes ?",
    "vgm": "Quelle est la valeur du VGM ?",
    "ccmh": "Quelle est la valeur de la CCMH ?",
    "tcmh": "Quelle est la valeur du TCMH ?"
}

# Construction de la liste Q&A
qas = []
for champ, question in questions.items():
    reponses = extraire_valeurs_sans_unite(champ, lignes_corrigees)
    if reponses:
        qas.append({
            "question": question,
            "answers": reponses
        })

# Contexte complet (texte brut)
context = "\n".join(lignes_corrigees)

# Structure JSON finale
json_data = {
    "id": "rapport_12",
    "context": context,
    "qas": qas
}

import json
print(json.dumps(json_data, ensure_ascii=False, indent=2))
```

```python
import re
import unicodedata
import cv2
import pytesseract
import json

# Config pytesseract
pytesseract.pytesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
image_path = "C:\\Users\\asus\\Desktop\\generation2\\generated_reports\\rapport_10.png"

zones_values = {
    'globules rouges': (550, 577, 53, 27),
    'hematocrite': (550, 604, 53, 27),
    'hemoglobine': (550, 632, 53, 28),
    'vgm': (550, 660, 53, 28),
    'tcmh': (550, 690, 53, 27),
    'ccmh': (550, 717, 53, 27),
    'leucocytes': (546, 745, 59, 27),
    'neutrophiles': (720, 822, 53, 27),
    'eosinophiles': (724, 849, 45, 27),
    'basophiles': (724, 876, 45, 28),
    'lymphocytes': (720, 905, 53, 28),
    'monocytes': (720, 935, 53, 28),
    'plaquettes': (525, 970, 72, 28),
}

# Fonction pour normaliser le texte OCR (enlever accents, minuscules)
def normaliser(texte):
    texte = unicodedata.normalize('NFD', texte)
    return ''.join(c for c in texte if unicodedata.category(c) != 'Mn').lower()

# Fonction OCR pour extraire tokens et bboxes
def ocr_image(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    data = pytesseract.image_to_data(gray, output_type=pytesseract.Output.DICT, lang='fra')
    tokens = []
    bboxes = []
    n_boxes = len(data['text'])
    for i in range(n_boxes):
        txt = data['text'][i].strip()
        conf = int(data['conf'][i])
        if txt == '' or conf < 0:
            continue
        # bbox format: [x0, y0, x1, y1]
        x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
        tokens.append(txt)
        bboxes.append([x, y, x + w, y + h])
    return tokens, bboxes

# Fonction pour extraire valeur dans une zone donnée par OCR + nettoyage
def extract_value_zone(image_path, bbox):
    from PIL import Image
    import pytesseract
    import re
    image = Image.open(image_path)
    x, y, w, h = bbox
    cropped = image.crop((x, y, x + w, y + h))
    custom_config = r'--oem 3 --psm 6 outputbase digits'
    text = pytesseract.image_to_string(cropped, config=custom_config)
    text = re.sub(r'[^0-9.,]', '', text)
    return text.strip()

# Charger tokens + bboxes OCR complet
tokens, bboxes = ocr_image(image_path)

# Extraire les valeurs dans les zones fixes (zones_values)
from PIL import Image
image = Image.open(image_path)
valeurs_zones = {}
for cle, bbox in zones_values.items():
    valeur = extract_value_zone(image_path, bbox)
    valeurs_zones[cle] = valeur

print("Valeurs extraites par zones :")
for k,v in valeurs_zones.items():
    print(f"{k} : {v}")

# Now assign NER labels to tokens for each value
# Format labels: B-KEY, I-KEY, O
labels = ["O"] * len(tokens)

def label_tokens_for_value(tokens, valeurs_zones, labels):
    # Pour chaque clé + valeur, on va essayer de trouver la valeur dans les tokens OCR
    for key, valeur in valeurs_zones.items():
        if not valeur:
            continue
        valeur_norm = normaliser(valeur)
        # Chercher la valeur dans la séquence tokens (concat token by token)
        for i in range(len(tokens)):
            # Construire une phrase à partir de i avec la taille du split valeur
            # On va essayer avec découpage naïf en tokens
            longueur_valeur = len(valeur_norm.split())
            phrase = ' '.join(normaliser(t) for t in tokens[i:i+longueur_valeur])
            if valeur_norm in phrase:
                # Label B- et I- sur la plage
                labels[i] = "B-" + key.upper().replace(' ', '_')
                for j in range(i+1, i+longueur_valeur):
                    if j < len(labels):
                        labels[j] = "I-" + key.upper().replace(' ', '_')
                break
    return labels

labels = label_tokens_for_value(tokens, valeurs_zones, labels)

# Préparer la sortie JSONL (exemple pour un seul document)
json_dict = {
    "id": "rapport_1",
    "tokens": tokens,
    "bboxes": bboxes,
    "labels": labels
}

# Sauvegarder dans un fichier jsonl
with open("rapport_1.jsonl", "w", encoding="utf-8") as f:
    f.write(json.dumps(json_dict, ensure_ascii=False) + "\n")

print("Fichier JSONL généré : rapport_1.jsonl")
```

```python
import re
import unicodedata
import cv2
import pytesseract
from difflib import get_close_matches
import os
import json

# ---------------- CONFIG ----------------
pytesseract.pytesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
image_dir = r"C:\Users\asus\Desktop\generation2\77\1\generated_reports"
output_json = "resultats_layoutlmv3.json"

unites_valides = {
    "g/dL", "pg", "µm³", "um³", "fL", "%", "10^3/uL", "10^6/uL",
    "mm³", "10*3/uL", "10*6/uL", "/mm³", "10*3/mm³", "10*6/mm³", "/ml", "g/dl", "Pg",
    "10*6/µl", "/µl", "/ul", "millions/mm", "M/mm³", "femtolites", "10⁶/mm³",
    "10p6/mm³", "10p3/mm³", "10 p 6/mm³", "10 p 3/mm³"
}

champs_cibles = [
    "leucocytes", "hématies", "plaquettes", "hémoglobine", "hématocrite",
    "neutrophiles", "lymphocytes", "eosinophiles", "basophiles", "monocytes",
    "globules rouges", "globules blancs", "vgm", "ccmh", "tcmh"
]

champs_sans_accents = [
    re.sub(r'[^a-z0-9\s]', '', unicodedata.normalize('NFD', c).lower())
    for c in champs_cibles
]

questions = {c: f"Quelle est la valeur des {c} ?" if c not in {"vgm","ccmh","tcmh"} 
             else f"Quelle est la valeur du {c} ?" for c in champs_cibles}

# ---------------- FONCTIONS UTILES ----------------
def normaliser(mot):
    mot = unicodedata.normalize('NFD', mot)
    return ''.join(c for c in mot if unicodedata.category(c) != 'Mn').lower()

def nettoyer_et_enlever_accents(texte):
    texte = unicodedata.normalize('NFD', texte.lower())
    return re.sub(r'[^a-z0-9\s]', '', ''.join(c for c in texte if unicodedata.category(c) != 'Mn'))

def est_unite_complexe(mot):
    pattern = r'^\d+p\d+\/mm[3³]$'
    pattern_general = r'^\d+[a-z]+\d+\/[a-z0-9/³*^]+$'
    return bool(re.match(pattern, mot.lower()) or re.match(pattern_general, mot.lower()))

def separer_chiffres_unites(texte):
    mots = texte.split()
    mots_corriges = []
    for mot in mots:
        if est_unite_complexe(mot):
            mots_corriges.append(mot)
        else:
            mot_sep = re.sub(r'(\d)([a-zA-Zµ/%])', r'\1 \2', mot)
            mot_sep = re.sub(r'([a-zA-Zµ/%])(\d)', r'\1 \2', mot_sep)
            mots_corriges.append(mot_sep)
    return ' '.join(mots_corriges)

def corriger_mot_ocr(mot):
    mot = mot.replace("'", "").replace("`", "")
    mot_clean = nettoyer_et_enlever_accents(mot)
    match = get_close_matches(mot_clean, champs_sans_accents, n=1, cutoff=0.7)
    return champs_cibles[champs_sans_accents.index(match[0])] if match else mot

def corriger_unite_intelligente(mot):
    mot_normalise = normaliser(mot)
    correspondances = get_close_matches(mot_normalise, [normaliser(u) for u in unites_valides], n=1, cutoff=0.6)
    if correspondances:
        index = [normaliser(u) for u in unites_valides].index(correspondances[0])
        return list(unites_valides)[index]
    return mot

def corriger_ligne_intelligente(ligne):
    ligne = separer_chiffres_unites(ligne)
    mots = ligne.split()
    mots_corriges = []
    for mot in mots:
        mot_corrige = corriger_mot_ocr(mot)
        if mot_corrige == mot:
            mot_corrige = corriger_unite_intelligente(mot)
        mots_corriges.append(mot_corrige)
    return ' '.join(mots_corriges)

# Toujours prendre la première valeur pour tous les champs
def extraire_valeurs_selon_champ(champ, ligne):
    ligne_sep = separer_chiffres_unites(ligne)
    valeurs = re.findall(r'\d+[.,]?\d*', ligne_sep)
    valeurs = [v.replace(',', '.') for v in valeurs]
    if not valeurs:
        return None
    return valeurs[0]

# ---------------- TRAITEMENT DES IMAGES ----------------
resultats = []

for filename in os.listdir(image_dir):
    if not filename.lower().endswith((".png", ".jpg", ".jpeg")):
        continue
    print(f"Traitement de {filename}...")
    image_path = os.path.join(image_dir, filename)
    image = cv2.imread(image_path)
    if image is None:
        print(f"⚠️ Impossible de lire {filename}, passage au suivant.")
        continue

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    data = pytesseract.image_to_data(gray, output_type=pytesseract.Output.DICT, lang='fra')
    n_boxes = len(data['text'])

    # Regroupement mots par ligne
    lines = {}
    tolerance_y = 15
    for i in range(n_boxes):
        mot = data['text'][i].strip()
        conf = int(data['conf'][i])
        if mot == '' or conf < 0:
            continue
        y, x = data['top'][i], data['left'][i]
        found = False
        for key in lines:
            if abs(key - y) <= tolerance_y:
                lines[key].append({
                    'word': mot,
                    'x': x,
                    'left': data['left'][i],
                    'top': data['top'][i],
                    'width': data['width'][i],
                    'height': data['height'][i]
                })
                found = True
                break
        if not found:
            lines[y] = [{
                'word': mot,
                'x': x,
                'left': data['left'][i],
                'top': data['top'][i],
                'width': data['width'][i],
                'height': data['height'][i]
            }]

    # Lignes corrigées
    lignes_corrigees = []
    lignes_mots_pos = []
    for y in sorted(lines):
        mots_tries = sorted(lines[y], key=lambda m: m['x'])
        ligne = " ".join(m['word'] for m in mots_tries)
        ligne_corrigee = corriger_ligne_intelligente(ligne)
        lignes_corrigees.append(ligne_corrigee)
        lignes_mots_pos.append(mots_tries)

    # Construction des QAs
    qas = []
    for champ, question in questions.items():
        for idx, ligne in enumerate(lignes_corrigees):
            if nettoyer_et_enlever_accents(champ) in nettoyer_et_enlever_accents(ligne):
                valeur = extraire_valeurs_selon_champ(champ, ligne)
                if valeur:
                    mots_pos = lignes_mots_pos[idx]
                    tokens = [m['word'] for m in mots_pos]
                    bboxes = [[m['left'], m['top'], m['left']+m['width'], m['top']+m['height']] for m in mots_pos]

                    # Trouver l’index du token réponse
                    start_idx = None
                    for i, tok in enumerate(tokens):
                        tok_norm = tok.replace(",", ".").replace(":", "").replace(" ", "").lower()
                        val_norm = valeur.replace(",", ".").replace(" ", "").lower()
                        if val_norm in tok_norm:
                            start_idx = i
                            break

                    if start_idx is None:
                        continue  # pas trouvé

                    qas.append({
                        "id": f"{filename}_{champ}",
                        "question": question,
                        "tokens": tokens,
                        "bboxes": bboxes,
                        "answers": [valeur],
                        "start_positions": [start_idx],
                        "end_positions": [start_idx]
                    })
                break  # un seul match par champ

    resultats.extend(qas)

# ---------------- SAUVEGARDE ----------------
with open(output_json, "w", encoding="utf-8") as f:
    json.dump(resultats, f, ensure_ascii=False, indent=2)

print(f"✅ Extraction terminée. Résultats sauvegardés dans {output_json}")
```
